```python
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter
```

```python
df = pd.read_csv("../src/data/processed/swiss_jobs_final.csv")
```

```python
# Extract and clean skills
all_skills = []
for skills in df["skills_extracted"].dropna():
    for skill in skills.split(","):
        skill = skill.strip()
        if skill:
            all_skills.append(skill)

skill_counts = Counter(all_skills)
top_skills = skill_counts.most_common(10)

# What technical skills are most demanded in the Swiss job market?

skills, counts = zip(*top_skills)

plt.figure()
plt.barh(skills, counts)
plt.xlabel("Number of job postings")
plt.title("Top 10 Most Demanded Skills in Switzerland")
plt.gca().invert_yaxis()
plt.show()
```

```python
# Which Swiss cantons have the highest number of job opportunities?

canton_counts = df["canton"].value_counts().head(10)

plt.figure()
plt.bar(canton_counts.index, canton_counts.values)
plt.xlabel("Canton")
plt.ylabel("Number of job postings")
plt.title("Top Swiss Cantons by Job Postings")
plt.show()
```


- R dominates the Swiss job market for data-related roles.
