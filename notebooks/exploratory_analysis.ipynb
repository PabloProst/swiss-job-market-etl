```python
import pandas as pd
from collections import Counter
```

```python
df = pd.read_csv("../src/data/processed/swiss_jobs_final.csv")
```

```python
all_skills = []

for skills in df["skills_extracted"].dropna():
    for skill in skills.split(","):
        cleaned_skill = skill.strip()
        if cleaned_skill:
            all_skills.append(cleaned_skill)

skill_counts = Counter(all_skills)

print(skill_counts.most_common(10))
```
